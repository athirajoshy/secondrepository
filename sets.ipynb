**Set**

A set is a collection which is unordered and unindexed. In Python sets are written with curly brackets or using set() keyword

```
a= {"apple", "kiwi", "cherry"}
print(a)
```

```
b={22,66,1,1,1}
b
```

```
  
# Creating a Set 
l=[1,2,3]
t=(2,4,42)
set1 = set(l) 
set2=set(t)
print(set1) #blank set
print(set2)
l2=list(set2)  
print(l2)
```

```
set1 = set(["example"]) 
set2=set('abcd')
print("Set with the use of String: ") 
print(set1)
print(set2)
```

```
#convert a list to set
set1 = set(["example", "of", "sets",1,1]) 

print(set1) 
```

A set contains only unique elements but at the time of set creation, multiple duplicate values can also be passed. Order of elements in a set is undefined and is unchangeable. Type of elements in a set need not be the same, various mixed up data type values can also be passed to the set.

```
set1 = set([1,'students', 4.0]) 
print("Set with different dtypes") 
print(set1) 
```

Once a set is created, you cannot** change ** its items, but you can  **add** new items.

**Add Items**

To add one item to a set use the **add()** method.

To add more than one item to a set use the **update()** method.

```
#add()

a.add("orange")

print(a)
```

```
#update()
a.update(("potato","rice"))
a
```

```
print(len(a))#find length of a set
```

**Remove Item**
To remove an item in a set, use the **remove()**, or  
**discard**  method

```
a.remove("rice")
a
```

```
a.remove("pineapple")#as item is not present it will generate error
a
```

```
a.discard(("apple","kiwi"))
a
```

```
a.discard("pineapple")#if item doesn't exist it will not raise an error
a
```

You can also use the **pop()**, method to remove an item, but this method will remove the last item. Remember that sets are unordered, so you will not know what item that gets removed.

The return value of the pop() method is the **removed item**.

```
#pop
b= {"apple", "banana", "cherry"}

x = b.pop()

print(x)

print(b)
```

```
b.clear()
print(b)
l=[3,4,5]
l.clear()
print(l)
```

```
#delete keyword is used to delete the set completely
thisset = {"apple", "banana", "cherry"}

del (thisset)

print(thisset)
```

**Join Two Sets**

The ** union()**  method returns a new set with all items from both sets

The **update()** method inserts the items in set2 into set1

both will exclude duplicates

```
set1 = set(["a", "b" , "c"])
set2 = set([1, 2, 3])

set3 = set1.union(set2)
print(set3)
```

```
x = {"a", "b" , "c"}
y = {1, 2, 3}

x.update(y)
print(x)
```
