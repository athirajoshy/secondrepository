```
#7 write a function to find sum of all numbers in a list
def sum_list(items):
    sum_numbers = 0
    for x in items:
        sum_numbers=sum_numbers+x #increment by 1--> +=
    return sum_numbers
print(sum_list([2,3,0]))
```

```
#8 write a program to find largest num of a list using function
def max_num_in_list( list ):
    max = list[ 0 ]
    for a in list:
        if a > max:
            max = a   #1,2
    
    return max
print(max_num_in_list([1, 2, -8, 0]))
```

```
 # 9 find the list of words that are longer than n from a given string using function
def test_string(n, str):
    longer_words = []
    test_split = str.split(" ") 
    print(test_split) 
    for x in test_split:
        if len(x) > n:
            longer_words.append(x)
    return longer_words
test_string(3, "hello i am a girl")
```

```
def factorial(num):
    if num == 1:
        return num
    else:
        return num * factorial(num - 1)
 
# User input 
num = int(input("Enter a Number: "))
 
# if input number is negative then return an error message
# elif the input number is 0 then display 1 as output
# else calculate the factorial by calling the user defined function
if num < 0:
    print("Factorial cannot be found for negative numbers")
elif num == 0:
    print("Factorial of 0 is 1")
else:
    print("Factorial of", num, "is: ", factorial(num))
```
