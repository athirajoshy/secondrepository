**Arrays**
A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers. 

```
#one dimensionalarray
import numpy as np 
n = np.array([1,2,3]) 
print(n)
```

**ARRAY ATTRIBUTES**

shape

ndim




```
#shape
import numpy as np 
a = np.array([[1,2,3],[4,5,6]])
print(a) 
print (a.shape)
print(a.ndim)
```

```
# resizing of ndarray 
import numpy as np 

a = np.array([[1,2,3],[4,5,6]]) 
print(a)
a.shape = (3,2) 
print(a)
print(a.ndim)
```

```
#reshape
import numpy as np 
a = np.array([[1,2,3],[4,5,6]]) 
print(a)
print(a.reshape(3,2)) 
#print(b)
```

```
#ndim
# this is one dimensional array 
import numpy as np 
a = np.arange(24)
print(a)
print(a.ndim)  
 
# now reshape it 
b = a.reshape(4,2,3) 
print(b)
print(b.ndim)
# b is having three dimensions
```
